```python
import requests

url = "https://img.wongnai.com/p/400x400/2025/02/13/ba9788b509d04562917006e8bc127546.jpg"
filename = r"E:\ProjectImage\image.jpg"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    with open(filename, 'wb') as f:
        f.write(response.content)
    print("Image downloaded successfully!")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import os

# Set up headless Chrome
options = Options()
# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
# options.add_argument("--no-sandbox")


failed_rows = []

# Start Chrome with auto-managed driver
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
base_dir = r"E:\ProjectImage"

for index, row in data.iterrows():
    url = row['menu_image']
    if url == '':
        continue
    restaurant = str(row['restaurant_name'])
    menu = str(row['menu_name'])
    
        # Use existing restaurant folder
    restaurant_folder = os.path.join(base_dir, restaurant)
    filename = f"{menu}.png"
    save_path = os.path.join(restaurant_folder, filename)

    try:
        driver.get(url)
        time.sleep(5)

        img = driver.find_element("tag name", "img")
        img.screenshot(save_path)
        print(f"✅ Saved: {filename}, Restaurant Name: {restaurant}")

    except Exception:
        failed_rows.append({
            "index": index,
            "restaurant_name": row['restaurant_name'],
            "menu_name": row['menu_name'],
            "menu_image": url
        })
        continue

driver.quit()
```

```python
df_missing = pd.read_csv('missing_uploads.csv')
base_path = Path("E:/ProjectImage")
uploaded_data = []
error_logs = []
for i, row in df_missing.iterrows():
    restaurant_name_original = row['restaurant_name']
    menu_name = row['menu_name']
    folder_path = base_path / restaurant_name_original
    file_path = folder_path / f"{menu_name}.png"
    cloudinary_folder = restaurant_name_original.replace("&", "and")
    try:
        result = cloudinary.uploader.upload(
        str(file_path),
        public_id=menu_name.replace("&", "and"),            # Keep original menu_name
        folder=cloudinary_folder,       # Clean only the folder
        use_filename=True,
        unique_filename=False
    )
        print(f"✅ Uploaded: {cloudinary_folder}/{menu_name} → {result['secure_url']}")
        uploaded_data.append({
            'restaurant_name': restaurant_name_original,
            'menu_name': menu_name,
            'cloudinary_url': result['secure_url']
        })
    except Exception as e:
        error_logs.append(file_path)
        continue
```
