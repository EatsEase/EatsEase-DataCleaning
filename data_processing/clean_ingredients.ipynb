```python
import pandas as pd


df = pd.read_csv('clean_menu - Allergies.csv')
df
```

```python
menu_df = pd.read_excel('clean_menu (2).xlsx')
menu_df
```

```python
# Convert df to a dictionary for lookup
ingredient_to_allergy = dict(zip(df['ingredients'], df['allergies']))
ingredient_to_allergy
```

```python
def find_allergies(ingredient_str):
    ingredients = [i.strip() for i in ingredient_str.split(',')]
    matched_allergies = [ingredient_to_allergy[i] for i in ingredients if i in ingredient_to_allergy]
    return ', '.join(sorted(set(matched_allergies))) if matched_allergies else ''

# Apply the function to menu_df
menu_df['matched_allergies'] = menu_df['ingredients'].apply(find_allergies)
```
